```python
# Convert the categorical data into numerical data
df['DHSREGNA']= df['DHSREGNA'].astype('category')
df['URBAN_RURA']= df['URBAN_RURA'].astype('category')
df['SOURCE']= df['SOURCE'].astype('category')
df['ADM1NAME']= df['ADM1NAME'].astype('category')
df['DHSREGNA'] = df['DHSREGNA'].cat.codes
df['URBAN_RURA'] = df['URBAN_RURA'].cat.codes
df['SOURCE'] = df['SOURCE'].cat.codes
df['ADM1NAME'] = df['ADM1NAME'].cat.codes
df[['URBAN_RURA','DHSREGNA','SOURCE']].head()
```
